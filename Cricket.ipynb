# ************* SOLVING DATA ANALYISIS QUESTIONS *************

# 1. Find out the carrier length for each player

```python
# Convert year columns to integers (handling non-numeric values)
df['Start_Year'] = pd.to_numeric(df['Start_Year'], errors='coerce')
df['End_Year'] = pd.to_numeric(df['End_Year'], errors='coerce')

# Now calculate the difference
df['Carrier_Length'] = df['End_Year'] - df['Start_Year']

# Optional: Fill NA values (if conversion failed)
df['Carrier_Length'] = df['Carrier_Length'].fillna(0)
df
```

# 2. Find Top 5 Performers

```python
df.sort_values(by='Runs', ascending = False).head(5)
```

# 3. Find the Players with Average over 50

```python
df[df['Average'] > 50]
```

# 4. Players with Highest Centuries

```python
df[['Player','Centuries']].sort_values(by='Centuries', ascending=False).head(10)
```

# 5. What is the average career length per country?

```python
df[['Country','Carrier_Length']].groupby('Country').agg({'Carrier_Length':'mean'})
```

# 6. Which player has the highest batting average?

```python
df[['Player','Average']].sort_values('Average', ascending=False).head(1)
```

# 7. Which players have the highest strike rates?

```python
df[['Player','Strike_Rate']].sort_values('Strike_Rate', ascending=False).head(1)
```

# 8. What is the distribution of centuries per player?

```python

# Group players by the number of 100s and count how many fall into each group
hundreds_distribution = df['Centuries'].value_counts().sort_index()

# Plotting the distribution
plt.figure(figsize=(10, 6))
hundreds_distribution.plot(kind='bar', color='skyblue')
plt.title("Distribution of 100s Scored by Players")
plt.xlabel("Number of 100s")
plt.ylabel("Number of Players")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# 9. What is the ratio of 50s to 100s for each player?

```python
df['50_to_100_ratio'] = df['Fifties'] / df['Centuries'].replace(0, pd.NA)

# Display relevant columns
df[['Player', 'Centuries', 'Fifties', '50_to_100_ratio']].sort_values(by='50_to_100_ratio', ascending=False)
```

# 10. Which players have the most consistent batting average over long careers.

```python
# Convert 'Zeros' column to numeric if needed
df['Ducks'] = pd.to_numeric(df['Ducks'], errors='coerce')

# Filter players with careers longer than 10 years
long_career_df = df[df['Carrier_Length'] > 10]

# Sort by highest average, and lowest ducks (optional consistency proxy)
consistent_players = long_career_df.sort_values(by=['Average', 'Ducks'], ascending=[False, True])

# Display top players
consistent_players[['Player', 'Carrier_Length', 'Average', 'Ducks']].head(10)
```

# 11. Which players have scored the most boundaries

```python

df['Fours'] = pd.to_numeric(df['Fours'].astype(str).str.replace('+', '', regex=False), errors='coerce')
df['Sixes'] = pd.to_numeric(df['Sixes'].astype(str).str.replace('+', '', regex=False), errors='coerce')


df['Total_Boundaries'] = df['Fours'] + df['Sixes']

most_boundaries = df[['Player', 'Fours', 'Sixes', 'Total_Boundaries']].sort_values(by='Total_Boundaries', ascending=False)


most_boundaries.head(10)
```

# 12. Who has the highest percentage of not outs per innings?

```python
df['Percentage'] = (df['Not_Outs'] / df['Innings']) * 100

percentage_not_outs = df[['Player', 'Not_Outs', 'Innings', 'Percentage']].sort_values(by='Percentage', ascending=False)

percentage_not_outs.head(10)
```

# 13. Which decade produced the most high-average player?

```python

# Step 1: Create 'Decade' column
df['Decade'] = (df['Start_Year'] / 10) * 10

# Step 2: Filter high-average players
high_avg_df = df[df['Average'] > 50]

# Step 3: Count how many high-average players started in each decade
decade_counts = high_avg_df['Decade'].vsort_values(ascending=False).head(10)

# Display
decade_counts_df
```

# 14. Is there a correlation between strike rate and batting average?

```python
# Step 1: Convert Strike_Rate to numeric
df['Strike_Rate'] = pd.to_numeric(df['Strike_Rate'], errors='coerce')

# Step 2: Drop rows with missing values in either Strike_Rate or Average
clean_df = df.dropna(subset=['Strike_Rate', 'Average'])

# Step 3: Compute correlation
correlation = clean_df['Strike_Rate'].corr(clean_df['Average'])
correlation

plt.figure(figsize=(8,6))
plt.scatter(clean_df['Strike_Rate'], clean_df['Average'], alpha=0.7, color='green')
plt.title('Correlation Between Strike Rate and Batting Average')
plt.xlabel('Strike Rate')
plt.ylabel('Batting Average')
plt.grid(True)
plt.show()
```

# 15. Whatâ€™s the average number of matches per year per player?

```python
# Calculate average matches per year
df['Carrier_Length'] = df['End_Year'] - df['Start_Year']
df['Matches_per_Year'] = df['Matches'] / df['Carrier_Length']

# View top players with highest match frequency
df[['Player', 'Matches', 'Carrier_Length', 'Matches_per_Year']].sort_values(by='Matches_per_Year', ascending=False)
```

# 16. Which country has the highest average player strike rate?

```python
avg_strike_rate = df.groupby('Country').agg({'Strike_Rate':'mean'})
ans = avg_strike_rate.sort_values(by='Strike_Rate', ascending= False).head(10)
ans
```

# 17. How many players have a highest score above 200?


```python
# Step 1: Clean 'Highest_Score' to remove '*' and convert to numeric
df['Highest_Score_Clean'] = pd.to_numeric(df['Highest_Score'].astype(str).str.replace('*', '', regex=False), errors='coerce')

# Step 2: Filter players who scored 200 or more
high_scorers_200 = df[df['Highest_Score_Clean'] >= 200]

# Step 3: Count them
num_high_scorers = high_scorers_200.shape[0]
print(f"Number of players who scored 200 or more: {num_high_scorers}")
```
